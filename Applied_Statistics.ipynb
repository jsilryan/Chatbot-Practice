#### Probability Density Functions

```python
from scipy.stats import norm
```

```python
# Parameters
mu = 50
sigma = 5

# Distribution
dist = norm(mu, sigma)

values = [value for value in range(30, 70)]
probabilities = [dist.pdf(value) for value in values]

pyplot.plot(values, probabilities)
pyplot.show()

# Peak of probability is close to the mean (50)
```

```python
# Cumulative Probability - Probability of values over the same range
# At the mean, we've covered about 50% of the values, and close to 100% after a value of 65 or 3 SD from the mean
sample_space = arange(-5, 5, 0.001)

cdf = norm.cdf(sample_space)

pyplot.plot(sample_space, cdf)
pyplot.show()
```

```python
# Mean = 50, SD = 5; data is covered majorly within 2 SD to the mean
# Percentage Point Function
low_end = dist.ppf(0.025)
high_end = dist.ppf(0.975)

print('Mid 95%% between %.1f and %.1f' % (low_end, high_end))
```

```python
from scipy.stats import chi2, t, pearsonr, ttest_ind, ttest_rel, f_oneway # Calculate chi-squared distribution
```

```python
sample_space = arange(0, 50, 0.01)
dof = 20

pdf = chi2.pdf(sample_space, dof)

pyplot.plot(sample_space, pdf)
pyplot.show()

# Expected value of distribution is just short of the value 20 in the sample space
# Most of density in Gaussian distribution lies between -1 and 1, and sum of squared random observations from the standard gaussian distribution would sum to just under the number of degrees of freedom = 20
# Not symmetric
```

```python
# See likelihood of chi-squared value around 20
cdf = chi2.cdf(sample_space, dof)

pyplot.plot(sample_space, cdf)
pyplot.show()
```

```python
seed(1)

# randn - generate random numbers with a mean of 0 and std of 1
# Multiply with my own std and add my own mean - get my range
data1 = 20 * randn(1000) + 100
data2 = data1 + (10 * randn(1000) + 50) # Noise added

print("Data 1: Mean:%.3f | Stdv: %.3f"% (mean(data1), std(data1)))
print("Data 2: Mean:%.3f | Stdv: %.3f"% (mean(data2), std(data2)))

pyplot.scatter(data1, data2)
pyplot.show()
```

```python
covariance = cov(data1, data2)
print(covariance)
```

```python
corr, _ = pearsonr(data1, data2)
print("Pearson's Correlation: %.3f"% corr)
```

```python
seed(1)

data1 = 5 * randn(100) + 50
data2 = 5 * randn(100) + 51

print("Data 1: Mean = %.3f | Stdv: %.3f"% (mean(data1), std(data1)))
print("Data 2: Mean = %.3f | Stdv: %.3f"% (mean(data2), std(data2)))

pyplot.scatter(data1, data2)
pyplot.show()
```

#### Student's T-test

```python
# Null hypothesis - mean of the 2 samples are equal
# Rejection - sufficient evidence that the means are different and distributions aren't equal
# ttest_ind => 2 samples as arguments and returns calculated statistic and P-value
# Test discovers differences in distribution btn 2 independent samples

# stat: The t-statistic, which measures the difference between means in relation to the variability of the samples.
# p: The p-value, which tells us the probability that the observed difference happened by random chance.

stat, p = ttest_ind(data1, data2)
print("Statistics = %.3f, p = %.3f"% (stat, p))

alpha = 0.05
if p > alpha:
  print("Same Distributions (Fail to Reject H0)")
else:
  print("Different Distributions (Reject H0)")
```

#### Paired Student's T-Test

```python
# 2 independent measures or evaluations of the same object
# Compare means of 2 data samples related in a way
# Paired samples - repeated measures -> Dependent
# Can't use student's T-test as the measures are not independent
# Simplified as it doesn't assume there is a variation in observations

# Null hypothesis - No difference in sample means
# H1 - There is a difference in sample means

data3 = 5 * randn(100) + 50
data4 = 4 * randn(100) + 50
data5 = 5 * randn(100) + 52

stat, p = ttest_rel(data3, data4)
print("Statistics = %.3f, p = %.3f"% (stat, p))

alpha = 0.05
if p > alpha:
  print("Same Distributions (Fail to Reject H0)")
else:
  print("Different Distributions (Reject H0)")
```

```python
stat, p = f_oneway(data3, data4, data5)
print("Statistics = %.3f, p = %.3f"% (stat, p))

alpha = 0.05
if p > alpha:
  print("Same Distributions (Fail to Reject H0)")
else:
  print("Different Distributions (Reject H0)")
```
